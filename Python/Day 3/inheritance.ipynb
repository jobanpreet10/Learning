```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Student(Person):
    def __init__(self, name, age, marks):
        # Added self here
        Person.__init__(self, name, age)
        self.marks = marks

class GradeLevel(Person): # Renamed 'Class' to avoid reserved keyword confusion
    def __init__(self, name, age, grade):
        # Added self here
        Person.__init__(self, name, age)
        self.grade = grade

class s1(Student, GradeLevel):
    def __init__(self, name, age, marks, grade, address):
        # Manually calling specific parents
        Person.__init__(self, name, age, marks)
        GradeLevel.__init__(self, name, age, grade)
        self.address = address

p1 = s1("Joban", 20, 90, "A", "Jammu And Kashmir")
print(f"Name: {p1.name}, Marks: {p1.marks}, Grade: {p1.grade}, Address: {p1.address}")
```
