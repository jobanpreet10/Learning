```python
class MobilePhone:
    def __init__(self,b,p):
        self.brand = b
        self.price = p
    
    def call(self):
        print(self.brand,"is making a call")

    def show_price(self):
        print("Price :",self.price)
    


phone1 = MobilePhone("Samsung",30000)
phone2 = MobilePhone("Nokia",100000)

phone1.call()
phone2.show_price()
```

```python
class BankAccount:
    def __init__(self,name,balance):
        self.name = name
        self.balance = balance

    def deposit(self,amount):
        self.balance += amount
        print(amount,"The amount is deposited")

    def withdraw(self,amount):
        if self.balance < amount:
            print("Insufficient Balance")
        else:
            self.balance -= amount
            print(f"{amount} is withdrawn")
    
    def show_balance(self):
        print(f"The {self.name} holds the account")
        print(f"The balance of {self.name} is {self.balance}")
    


user1 = BankAccount("Joban",400000)
user1.deposit(600000)
user1.withdraw(500)
user1.show_balance()

        
```

```python
class Order:
    def __init__(self,customer_name):
        self.customer_name = customer_name
        self.products = []
        self.total = 0

    def add_products(self,product_name,price):
        self.products.append(product_name)
        self.total += price
    def apply_discount(self,percentage):
        discount = self.total*(percentage/100)
        self.total -= discount
    
    def generate_invoice(self):
        print(f"The customer is {self.customer_name}")
        print(f"The products bought are {self.products}")
        print(f"The final amount to be paid after the discount is {self.total}")


order1 = Order("Joban")
order1.add_products("Laptop",250000)
order1.add_products("Phone",150000)
order1.apply_discount(10)
order1.generate_invoice()

```

```python
class Product:
    def __init__(self,name,price):
        self.name = name
        self.price = price
    
    def __str__(self):
        return f"Product: {self.name} | Price: ${self.price}"
    
    def __add__(self,other1):
        return Product ("Combined",self.price + other1.price)
    


p1 = Product("Laptop",40000)
p2 = Product("Mobile",2000)
p3 = Product("hi",30000)

total = p1 + p2 + p3

print("Total :",total)

print(p1)
```

```python

def my_decorator(func):
    
    def wrapper():
        print("Before Function")
        func()
        print("After function")
    return wrapper

@my_decorator
def greet():
    print("Hello User")

greet()
```

```python
# Logging Decorator

def request_logger(func):
    def wrapper():
        print("Request started")
        func()
        print("Request Completed")
    return wrapper

@request_logger
def get_customer_data():
    print("Fetch customer data from datbase")

@request_logger
def place_order():
    print("Order Placed Sucessfully")

get_customer_data()

place_order()

```
