```python
## Counting the ferquencies of the numbers

lis = [1,2,1,13,3,2,3,4,3,2,3]
dic = {}

for i in lis:
    dic[i] = dic.get(i,0) + 1

print(dic)

```

```python
dic = [{"Mum":1},{"Mum":2},{"Del":2},{"Mum":10},{"Rak":20}]

cnt = {}

for item in dic:
    for i in item:
        cnt[i] = cnt.get(i,0)+1
    

print(cnt)
```
