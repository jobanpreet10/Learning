```python
import logging
import csv 
import os
file = "student_info.log"

logging.basicConfig(
    filename =file,
    level = logging.INFO,
    format ="%(asctime)s - %(levelname)s - %(message)s",
    datefmt="%Y:%m:%d  %H:%M:%S"
)

def log_to_csv(logs_file):
    with open(file ,'r') as log, open("student_data.csv",'a',newline='') as csvf:
        writer = csv.writer(csvf)
        writer.writerow(["date","time","name","age"])

        for line in log:
            # Split date, time, and message
            try:
                date_time, _, message = line.partition(" - INFO - ")
                date, time = date_time.split(" ")

                
                # Extract name and age from **your actual log message**
                # Message format: "The student name is John and age is 21."
                name_part = message.split("The student name is ")[1].split(" and age")[0]
                age_part = message.split("age is ")[1].replace(".", "")

                writer.writerow([date,time,name_part,age_part])

            except Exception as e:
                print("Skipping the line due to the parsing the issue")

        
    
while(True):
    choice = int(input("1 => Add the student\n" 
                        "2 => Create the csv file for the input data "
                        "\n 3 => Exit"))
    if choice == 1:
        name = input("Enter the name: ")
        age = int(input("Enter the age: "))
        logging.info(f"The student name is {name} and age is {age}.")

    elif choice == 2:
        log_to_csv(file)
        print("CSV Created sucessfully.")

    else:
        break


    
```
