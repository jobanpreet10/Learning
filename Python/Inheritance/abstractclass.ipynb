```python
from abc import ABC, abstractmethod
```

```python
class Animal(ABC):
    @abstractmethod
    def speak():
        pass

class dog(Animal):
    def speak(self):
        return("Bow Bow")

d1 = dog()
(d1.speak)
```

```python
from abc import ABC, abstractmethod
class Animal(ABC):
    @property
    @abstractmethod
    def species(self):
        pass 

class Dog(Animal):
    @property
    def species(self):
        return "Canine"

# Instantiate the concrete subclass
dog = Dog()
print(dog.species)
```
