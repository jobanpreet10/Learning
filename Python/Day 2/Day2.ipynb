```python
orders = []
customer_ids = set()
REOPRT_FILE = "order_report.txt"

def add_order(order_id,customer_id,product,price,quantity):
    order = {
        "order_id" : order_id,
        "customer_id" : customer_id,
        "product" : product,
        "price" : price,
        "quantity" : quantity
    }
    orders.append(order)
    customer_ids.add(customer_id)


def calculate_bill(price,quantity,tax_rate = 0.18, discount_rate = 0.10):
    subtotal = price*quantity
    tax = subtotal*tax_rate
    discount = subtotal + discount_rate if subtotal > 1000 else 0
    total = subtotal + tax - discount

    return subtotal,tax,discount,total


def process_orders():
    totals = []
    for order in orders:
        subtotal,tax,discount,total = calculate_bill(
            order["price"],order["quantity"]
        )
        
        totals.append(total)
        print(order["order_id"],total)
    
    return totals

def save_report(totals):
    with open(REOPRT_FILE,"w") as file:
        for i in range(len(orders)):
            file.write(f"{orders[i]['order_id']} : {totals[i]}\n")

        file.write("\n Highest : " + str(max(totals)))
        file.write("\n Lowest : " + str(min(totals)))


add_order(101,"C01","Laptop",5000,1)
add_order(102,"C02","Mobile",15000,2)
totals = process_orders()
save_report(totals)
```

```python
# Case Study 
# Accept region wise sales 





list_of_dicts = [
    {'item': 'apple', 'quantity': 10, 'price': 1.50},
    {'item': 'banana', 'quantity': 25, 'price': 0.50},
    {'item': 'apple', 'quantity': 15, 'price': 1.50},
    {'item': 'banana', 'quantity': 10, 'price': 0.50}
]

def sum_key_values(data_list, key_to_sum):
    total = 0
    for d in data_list:
        if key_to_sum in d:
            total += d[key_to_sum]
    return {f'total_{key_to_sum}': total}

# Example usage:
total_quantity = sum_key_values(list_of_dicts, 'quantity')
print(total_quantity)

total_price = sum_key_values(list_of_dicts, 'price')
print(total_price)
```

```python
sales = []
for i in range(2):
    dicti = {}
    region = input("Enter the region name:")
    sale = int(input("Enter the sale of the region:"))

    dicti[region] = sale
    sales.append(dicti)


print(sales)
def sum_key_value(sale):
   
    total = {}
    for d in sale:
        for reg,sal in d.items():
            total[reg] = total.get(reg,0)+sal

    return total
   


final_regions = sum_key_value(sales)

total_sale = 0
n = 0
for val in final_regions.values():
    total_sale += val
    n += 1

avg_sale = total_sale/n
min_sale = min(final_regions.values())
max_sale = max(final_regions.values())

print("Max sales is :" ,max_sale)
print("Min sales is :" ,min_sale)
print("Avg sales is :" ,avg_sale)
print("total  sales is :" ,total_sale)

file_name = "sales_report.txt"

with open(file_name,"w") as f:
    
    f.write(f"Max Sales :{max_sale}\n")
    f.write(f"Min sales is :{min_sale}\n")
    f.write(f"Avg sales is :{avg_sale}\n")
    f.write(f"total  sales is :{total_sale}")



print("Report Generated Sucessfully")
```
