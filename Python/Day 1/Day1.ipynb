```python
n = int(input("Enter the number of the students"))
students = []

for i in range(n):
    name = input("Name:")
    roll_no = input("roll_no:")
    mark = input("Marks:").split(",")
    marks = [int(m) for m in mark]
    print(marks)
    total = sum(marks)
    avg = total/len(marks)

    if avg > 85:
        grade = 'A'
    elif avg >= 70:
        grade = 'B'
    elif avg >= 50:
        grade = 'C'
    else:
        grade = 'D'

    students.append((name,roll_no,marks,total,avg,grade))

for s in students:
    print(s)
    
```
