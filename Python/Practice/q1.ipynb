```python
n = int(input("Enter the no of student:"))
lis = []
for i in range (n):
    name = input()
    age = int(input())
    fee = float(input())
    premium = input()

    if fee == 0:
        break
    elif age < 18:
        continue
    elif premium == 'yes':
        fee = fee + (fee*.2)
    elif age > 60:
        fee = fee - (fee*.1)
    dic = {'name':name,'fee':fee,'premium' : premium}
    # val = [name,fee,premium]
    # lis.append(val)
    lis.append(dic)


print(f"Total students: {len(lis)}")
rev = 0
for temp in lis:
    rev += temp[1]
print(f"Total revenue {rev}")

print("Premium Students:")
for temp in lis:
    if temp[2] == 'yes':
        print(f"{temp[0]}")

for temp in lis:
    print(f"Name {temp[0]} Fee:{temp[1] :2f}")


    
```
