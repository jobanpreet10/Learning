```python
import pandas as pd
```

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

#load data into a Series object:
df = pd.Series(data)

print(df) 

# in this case calories act as index  and the list is the value 
```

```python

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)

print(df) 
```

```python
# get the sepecific rows  by giving the list of indexes
print(df.loc[[0,2]])
```

```python
# can change indexes by giving your names 
df_names = pd.DataFrame(data, index = ["day1","day2","data3"])
```

```python
print(df_names)
```

```python
# check if iloc works in named indexes

import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35]
}

df = pd.DataFrame(data, index=["a", "b", "c"])
print(df)



# you can  also select the specific column


```

```python
df.iloc[0]
```

```python
df.iloc[1,1]
```

```python
# iloc works both for named and unnamed indexes 
# so it is better to use the iloc
```
