```python
import pandas as pd
```

```python
a = [1,2,3,4,5]
```

```python
myseries = pd.Series(a)
```

```python
print(myseries[0])
```

```python
# adding the label name to the series

myseriesLables = pd.Series(a,["a","b","c","d","e"])
```

```python
print(myseriesLables)
```

```python
# key value pair like dictionary
calories = {"day1":300,"day2":400}
myseriesDict = pd.Series(calories)
```

```python
print(myseriesDict)
```

```python
#To select only some of the items in the dictionary, use the index argument and 
# specify only the items you want to include in the Series.

myseriesDict1 = pd.Series(myseriesDict,index=["day2"])
print(myseriesDict1)
```
